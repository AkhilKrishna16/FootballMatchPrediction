```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(2, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

Rolling averages indicate that the average is updated as new data points become available. It always considers a fixed number of the most RECENT observations. 

In this case, '2' is the window size for how large we want our calculations to be. 'closed' is the consideration of what we should exclude in our data -- 'left' means exclude the last point.

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols)) # this applies one function to each team
```

```python
# new predictions

def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] > '2022-01-01']
    rf.fit(train[predictors], train['target'])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted=preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

We need to combine the home and away teams together. For instance, for the Wolverhampton Wanderers, they always have the team as the full name but whenever they are the opponent, they are listed as the Wolves. This can cause fluxes in predictions as well as the data, which we can fix. 

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key
    
map_values = {
    "Brighton and Hove Albion": "Brighton",
    "Manchester United": "Manchester Utd",
    "Newcastle United": "Newcastle Utd",
    "Tottenham Hotspur": "Tottenham",
    "West Ham United": "West Ham",
    "Wolverhamptom Wanderers": "Wolves"
}

mapping = MissingDict(**map_values)
```

```python
mapping['West Ham United']
```

```python
combined['new_team'] = combined["team"].map(mapping)
```

```python
combined
```

```python
merged = combined.merge(combined, left_on=["date", "new_team"], right_on=["date", "opponent"])
```

This is combining the 'combined' dictionary and itself. In other words, we are trying to see if there are any matches between 'new_team' and 'opponent' itself, having its result. 

```python
merged
```

```python
merged[(merged['predicted_x'] == 1) & (merged["predicted_y"] == 0)]["actual_x"].value_counts()
```

Filtering Rows:

merged['predicted_x'] == 1: This part checks each value in the 'predicted_x' column of the DataFrame merged to see if it equals 1. It creates a boolean mask where True indicates that the condition is met (i.e., 'predicted_x' equals 1) and False otherwise.

merged["predicted_y"] == 0: Similarly, this part checks each value in the 'predicted_y' column of merged to see if it equals 0. It also creates a boolean mask.

(merged['predicted_x'] == 1) & (merged["predicted_y"] == 0): Here, both boolean masks are combined using the & operator. This operation returns True only where both conditions are met simultaneously, and False otherwise. In other words, it finds rows where 'predicted_x' equals 1 and 'predicted_y' equals 0.

Selecting Columns:

merged[(merged['predicted_x'] == 1) & (merged["predicted_y"] == 0)]["actual_x"]: After applying the boolean indexing with the combined conditions, this part selects only the 'actual_x' column from the resulting DataFrame.
Counting Unique Values:

.value_counts(): Finally, .value_counts() calculates the frequency of each unique value in the 'actual_x' column selected in the previous step. It returns a Series where the unique values in 'actual_x' are the index, and the corresponding counts are the values.
